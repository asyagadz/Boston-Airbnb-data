```python
data = listing[['price', 'neighbourhood_cleansed', 'property_type', 'room_type','accommodates', 'bathrooms', 'bedrooms',
         'beds','bed_type','amenities','square_feet', 'cancellation_policy', 'host_since', 'host_response_time',
         'host_response_rate','host_acceptance_rate','host_is_superhost', 'host_identity_verified', 'number_of_reviews',
         'reviews_per_month','review_scores_accuracy','review_scores_cleanliness','review_scores_checkin',
         'review_scores_communication','review_scores_location','review_scores_value']].copy()
```

```python
data.shape
```

```python
# change name column
data = data.rename(columns={'neighbourhood_cleansed': 'neighbourhood'})
# host_response_rate, from % to float
data['host_response_rate'] = data['host_response_rate'].replace('[\%,]', '', regex=True).astype(float)

# host_acceptance_rate, from % to float
data['host_acceptance_rate'] = data['host_acceptance_rate'].replace('[\%,]', '', regex=True).astype(float)
```

```python
def DaysPassed(col, df):
    # convert to datetime dtype
    df[col] = pd.to_datetime(df[col])
    
    # date when the dataset was scrapped
    last_scraped = pd.to_datetime("2016-09-07")
    
    # return the difference between the 2 dates in days
    return (last_scraped - df[col]).dt.days 
```

```python
# from date to days since
data['host_since'] = DaysPassed('host_since', data)
```

```python
# Amenities, clean
data['amenities'] = data['amenities'].replace("{", '', regex=True)
data['amenities'] = data['amenities'].replace("}", '', regex=True)
data['amenities'] = data['amenities'].replace("\"", '', regex=True)

# Convert to dummies
A = data['amenities'].str.get_dummies(sep= ",")
A = A.drop('translation missing: en.hosting_amenity_49', 1)
A = A.drop('translation missing: en.hosting_amenity_50', 1)

A.columns
```

```python
# drop amenities columns
data = data.drop('amenities', 1)
```

```python
# remove dollar sign and convert to float
data['price'] = data['price'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# create variable of interest price per accommodate (ppa)
data.insert(loc=0, column='ppa', value=data['price']/data['accommodates'])
# drop price and accommodate
data = data.drop(columns=['price', 'accommodates'])
```

```python
# delete square feet 
data = data.drop(columns=['square_feet'])
```

```python
# show number of null values per column in numerical variables
numerical = data.select_dtypes(include=['float64', 'int64'])
numerical = numerical[numerical.columns[1:15]]
numerical.columns
data[numerical.columns].isnull().sum()
```

```python
# fill nulls with median in numerical variables
for col in numerical.columns:
    m = data[col].median()
    data[col] = data[col].fillna(m)
```

```python

# verify that there are not nulls 
data.isnull().sum()
```

```python
# show the null values in categorical variables
cat = data.select_dtypes(exclude=['float64', 'int64'])
data[cat.columns].isnull().sum()
```

```python

# fill nulls with the mode in the categorical features
for col in cat.columns:
    mode = data[col].mode()[0]
    data[col] = data[col].fillna(mode)
```

```python
# no nulls in the whole dataset
print('number of null values in the whole dataset:', data.isnull().sum().sum())
```

```python
# select the categorical variables and create dummy
categorical = data[cat.columns]
dummies = pd.get_dummies(categorical)
dummies.head()
```

```python
# drop the categorical features from data
data = data.drop(columns=cat.columns)
# concatenate with dummies
data = pd.concat([data, dummies], axis=1)
data.head()
```

```python
data.shape
```

```python
# concatenate data with amenities
result = pd.concat([data, A], axis=1)
result.head()
```

```python
data=result.copy()
```

```python
# keep only the columns with true values
data = data.drop(['host_is_superhost_f', 'host_identity_verified_f'], axis=1)
```

```python
# prepare y and X
y = data['ppa']
X = data.drop('ppa', axis=1)

# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=333)

# create the model, fit and predict
rf = RandomForestRegressor(n_estimators = 500, max_depth = 10, random_state=333)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

# evaluate test set
mse_test = mean_squared_error(y_test, y_pred)
rmse_test = np.sqrt(mse_test)
print('RMSE test set:', round(rmse_test,3))

# check the error on the train set
y_pred_train = rf.predict(X_train)
mse_train = mean_squared_error(y_train, y_pred_train)
rmse_train = np.sqrt(mse_train)
print('RMSE train set:', round(rmse_train,3))

# prepare data for plot
d = {'y_test': y_test.values, 'y_pred': y_pred}
df = pd.DataFrame(d)
df = df.sort_values('y_test')
df = df.reset_index(drop=True)

# plot true (test) vs predicted values
df.plot()
plt.ylabel('price')
plt.xlabel('n')
plt.title('true vs predicted values random forest')
#plt.savefig('figures/6_true_pred.png', bbox_inches='tight')
plt.show()
```
